for vai executar as instruções através de uma sequência conhecida for (i = 0; i < 10; i++) - percorrer listas, vetores, executar um número definido x de ações e outros.

````python
def impar_ou_par(n):
  if n % 2 == 0:
    print(f'{n} é par')
  else:
    print(f'{n} é ímpar')
````

````python
impar_ou_par(51)
````

````python
impoupar = lambda n: print(f'{n} é par') if n % 2 == 0 else print(f'{n} é impar') #lambda, argumento, retorno do argumento, condição
````

````python
impoupar(50)
````
